## Qué géneros tienen mejores reviews de media?

```python
df.head()
```

```python
ventas_por_genero.head()
```

```python
ventas_por_genero = df_concat.iloc[:, 9:].sum()
```

```python
lista_generos = ['action', 'adventure', 'rpg', 'simulation', 'strategy', 'racing',
       'free_to_play', 'early_access']
```

```python
def mejores_review(lista_generos):
    review_generos = []
    
    for genero in lista_generos:
        categoria = genero
        game = df_concat.loc[(df_concat[genero] == True)]
        total1 = game["negative_ratings"].sum()
        total2 = game["positive_ratings"].sum()
        total2
        total = total2 - total1
        review_generos.append(total)
    return review_generos
```

```python
more_dear = mejores_review(lista_generos)
```

```python
mejores_review_final = {"genero": lista_generos,
                        "review": more_dear}
```

```python
df_secundario = pd.DataFrame(mejores_review_final)
```

```python
df_secundario
```

```python
fig= plt.figure(figsize= (16,10))

g = sns.barplot(x = df_secundario["genero"],
                y = df_secundario["review"])


for index, value in enumerate(df_secundario["review"]):
    g.text(index, value, str(value), color='black', ha="center")

    
fig.suptitle("Géneros más queridos")
g.set_ylabel("Reviews positivas")
g.set_xlabel("Generos")



plt.show()
```

```python
fig= plt.figure(figsize= (16,10))

g = sns.barplot(x = ventas_por_genero.index,
                y = ventas_por_genero.values)



for index, value in enumerate(ventas_por_genero.values):
    g.text(index, value, str(value), color='black', ha="center")

g.set_title("Juegos mas vendidos por genero")
g.set_ylabel("Ventas")
g.set_xlabel("Generos")



plt.show()
```

## En qué mes, desde que tenemos datos, se han publicado más juegos?

```python
df.head()
```

```python
import calendar
df['month_release'] = df['release_date'].dt.month#.apply(lambda x: calendar.month_name[x])
```

```python
df
```

```python
meses_lanzamiento = df['month_release'].value_counts().sort_index()
```

```python
meses_lanzamiento
```

```python
fig= plt.figure(figsize= (16,10))

g = sns.barplot(x = meses_lanzamiento.index,
                y = meses_lanzamiento.values)


for index, value in enumerate(meses_lanzamiento.values):
    g.text(index, value, str(value), color='black', ha="center")

    
fig.suptitle("Meses con mas lanzamientos")
g.set_ylabel("Cantidad de lanzamientos")
g.set_xlabel("Meses")



plt.show()
```

```python
import pandas as pd

# Suponiendo que los datos están en un dataframe llamado df

# Obtener las columnas de género
columnas_genero = ['action', 'adventure', 'rpg', 'simulation', 'strategy', 'racing', 'free_to_play', 'early_access']

# Utilizar la función melt para convertir las columnas de género en una sola columna
df_melted = df_concat.melt(id_vars=['name', 'release_date', 'developer', 'achievements', 'positive_ratings', 'negative_ratings', 'avg_hours_per_user', 'price', 'sells', 'month_release'], 
                    value_vars=columnas_genero, 
                    var_name='genero', 
                    value_name='pertenece')

# Filtrar solo los juegos que pertenecen al género
df_filtrado = df_melted[df_melted['pertenece']]

# Eliminar la columna 'pertenece' ya que los valores son todos True
df_filtrado = df_filtrado.drop('pertenece', axis=1)
```

```python
df_filtrado
```

```python
df_agrupado = df_filtrado.groupby(['genero', 'month_release'])['avg_hours_per_user'].sum().reset_index()
```

```python
df_promedio = df_filtrado.groupby('genero')['avg_hours_per_user'].mean()
```

```python
df_promedio
```

```python
fig= plt.figure(figsize= (16,10))

g = sns.barplot(x = df_promedio.index,
                y = df_promedio.values)

    
fig.suptitle("Meses con mas lanzamientos")
g.set_ylabel("Cantidad de lanzamientos")
g.set_xlabel("Meses")



plt.show()
```
